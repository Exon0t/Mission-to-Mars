```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import re
import pandas as pd 
```

```python
from IPython.core.display import HTML
HTML('''''')
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.artscomplex.com/events/all"
browser.visit(url)
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
calendar = html_soup.find('div', class_ = "tl-calendar")#get the date
dates = calendar.find_all('div', class_ = re.compile("tl\-date day_\d+ count_\d+ hasEvent"))
new_event = []
event_location = {}
my_calendar = {
    "event": [],
    "location":[],
    "date":[]
}
for d in dates:
    date = d["data-fulldate"]
    events = d.find_all('div',class_="event_item")
    print(date)
    for x in events:
        html = browser.html
        html_soup = soup(html, 'html.parser')
        event = x.h3.text
        cl = x.h3.a['href']
        #if its a new event
        if event not in new_event:
            new_event.append(event)
            #click
            browser.visit(cl)
            html = browser.html
            html_soup = soup(html, 'html.parser')
            location = html_soup.find("li", class_= "item sidebar_event_venue").text.replace("\n", "")
            browser.back()
            event_location[event] = location.replace("Venue", "")
        print(f"{event} ------ {event_location[event]}")
        my_calendar["event"].append(event)
        my_calendar["location"].append(event_location[event])
        my_calendar["date"].append(date)
    print("-----------------")
    
# print(calendar.div.h3.prettify())
# calendar.div
```
